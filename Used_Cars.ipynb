#### Creating Brand Mean price and brand milage

```python
autos.brand.value_counts()
```

```python
intersting_brands=autos.brand.value_counts()[0:22].index
intersting_brands
# I chose up too Kia
```

```python
brand_mean_price={}

for b in intersting_brands:
    brand_mean_price[b]=int(autos['price'][autos['brand']==b].mean())
brand_mean_price
y=pd.Series(brand_mean_price)
y
```

```python
brand_mean_milage={}
for b in intersting_brands:
    brand_mean_milage[b]=int(autos['kilometer'][autos['brand']==b].mean())
x=pd.Series(brand_mean_milage)
x
```

```python
brand_means=pd.DataFrame([y,x],index=['price','milage']).T.sort_values('price',ascending=False)
brand_means
```
